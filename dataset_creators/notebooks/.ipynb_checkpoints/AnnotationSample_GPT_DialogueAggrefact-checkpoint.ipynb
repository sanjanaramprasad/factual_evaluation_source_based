```python
import pandas as pd
import openai
```

```python
import os


from nltk import word_tokenize, sent_tokenize
from tenacity import (
    retry,
    stop_after_attempt,
    wait_random_exponential,
)  # for exponential backoff


from transformers import GPT2Tokenizer
class GPTInference():
    def __init__(self):
        OPENAI_API_KEY = "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"
        os.environ["OPENAI_API_KEY"] = OPENAI_API_KEY
        openai.api_key = os.getenv("OPENAI_API_KEY")

    @retry(wait=wait_random_exponential(min=1, max=60), stop=stop_after_attempt(6))
    def get_chatgpt_response(self, prompt):
        # query = f'Article: {doc}\n{instruction}\n{prompt}'
        response = openai.ChatCompletion.create(model="gpt-4-32k-0613",
                                       messages=[
                                           
                        {"role": "user", "content": f'{prompt}'},   
                        ], 
                        )
#         print(query)
#         print(response)
        return response['choices'][0]['message']['content']
```

```python
df = pd.read_csv('/home/sanjana/factual_evaluation_source_based/datasets/sota_annotations/dialogue_aggrefact.csv')
len(df)
```

```python
df_annotations = {'DocID': [], 
                  'Dialogue': [],
                  'origin': [],
                 }
                  
                  
for idx, row in df.iterrows():
    docid = row['DocID']
    dialogue = row['Dialogue']
    origin = row['origin']

    if docid not in df_annotations['DocID']:
        df_annotations['DocID'].append(docid)
        df_annotations['Dialogue'].append(dialogue)
        df_annotations['origin'].append(origin)
    else:
        if dialogue not in df_annotations['Dialogue']:
            print(docid)
df_annotations = pd.DataFrame(df_annotations)
```

```python
from tqdm import tqdm
def make_gpt4_summaries(df_annotations):
    gpt_model = GPTInference()
    df_dict = {
        'DocID': [],
        'Dialogue': [],
        'Model': [],
        'Summary': [],
        'origin': []
        
    }
    for idx, row in tqdm(df_annotations.iterrows(), total=df_annotations.shape[0]):
        docid = row['DocID']
        dialogue = row['Dialogue']
        origin = row['origin']
        instr = 'Generate a summary of the dialogue snippet provided above'
        prompt = f'{dialogue}\n{instr}\nSummary:'
        summary = gpt_model.get_chatgpt_response(prompt)
        # print(prompt)

        df_dict['DocID'].append(docid)
        df_dict['Dialogue'].append(dialogue)
        df_dict['Model'].append('gpt4-32k-0613')
        df_dict['Summary'].append(summary)
        df_dict['origin'].append(origin)
        
    
    df = pd.DataFrame(df_dict)
    # df.to_csv(out_file)
    return df
```

```python
df_annotations_gpt = make_gpt4_summaries(df_annotations)
```

```python
df_annotations_gpt.to_csv('/home/sanjana/factual_evaluation_source_based/annotations/round1_aggrefact/gpt4_summaries_short.csv')

```
