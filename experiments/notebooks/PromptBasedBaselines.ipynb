```python
import pandas as pd
from GPTModel import GPTInference
```

```python
sota_path = '/home/sanjana/factual_evaluation_source_based/datasets/sota_annotations/dialogue_aggrefact.csv'
    
def read_data(path):
    return pd.read_csv(path)

def next_summary_source_pair(): 
    df = read_data(sota_path)
    df = df[df['Model'] != 'human_ref']
    dialogues = list(df['Dialogue'].values)
    summaries = list(df['Summary'].values)
    models = list(df['Model'].values)
    for dlg, summ, mdl in list(zip(dialogues, summaries, models)):
        yield dlg, summ, mdl

                   
```

```python
def pred_label_prompt():
    gpt_model = GPTInference()
    all_answers = []
    label_instruction = 'Decide whether the summary is consistent with the dialogue. Note consistency means all the information in the summary can be found or inferred from the dialogue. A consistent summary does not contain contradictory or new information.'
    for dlg, summ, model in next_summary_source_pair():
        prompt = f'{instruction}\nDialogue: {dlg}\nSummary: {summ}\n Answer(yes or no):'
        answer = gpt_model.get_chatgpt_response(prompt)
        all_answers += [answer]
    return all_answers 
```
