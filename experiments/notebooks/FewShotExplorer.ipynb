```python
sota_path = '/home/sanjana/factual_evaluation_source_based/datasets/sota_annotations/RefMatters_SAMSum_train.csv'
    
def read_data(path):
    return pd.read_csv(path)
    
```

```python
df = read_data(sota_path)
df.head()[:1]
```

```python
list(set(df.keys()))
```

```python

df_errors = df[df['w/ Error'] == 1]
```

```python
from random import sample 
error_types = ['PredE', 'SubjObjE', 'OutE', 'CorefE', 
               'GramE', 'CircE', 'LinkE', 'OthE']
df_acc = []
for err_type in error_types:
    df_err = df_errors[df_errors[err_type] == 'yes']
    df_err_sample = df_err.sample(1)
    df_acc.append(df_err_sample)
```

```python
df_acc = pd.concat(df_acc)
```

```python
for idx, row in df_acc.iterrows():
    dialogue = row['Dialogue']
    summary = row['Summary']
    print( dialogue)
    print(summary)
    for err_type in error_types:
        if row[err_type] == 'yes':
            print(err_type)
            print(row[err_type])
            print(row[f'{err_type}_text'])
    print('*'*13)
            
```
